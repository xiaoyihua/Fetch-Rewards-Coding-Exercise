```python
# add commas between each json object by splitting into individual jsons and then joining with commas

# File paths
input_file_path = 'jsons/receipts.json'  # Replace with your actual file path
output_file_path = 'formatted_receipts.json'  # Output file path

# Read the original JSON file
with open(input_file_path, 'r') as file:
    data = file.read()

# Split the data into individual JSON objects
json_objects = data.splitlines()

# Remove any empty lines and join with commas
json_objects = [obj for obj in json_objects if obj.strip()]  # Filter out empty lines
formatted_json = '[' + ',\n'.join(json_objects) + ']'

# Write the formatted data to a new file
with open(output_file_path, 'w') as output_file:
    output_file.write(formatted_json)
```

```python
# finally opening the file
with open('jsons/formatted_receipts.json') as f:
    receipts_data = json.load(f)
```

```python
# Convert to DataFrame
receipts_df = pd.DataFrame(receipts_data)
```
